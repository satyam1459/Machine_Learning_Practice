**Linear Regression Stats Model**

```python
import pandas as pd
df = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
# Median House Price
y = df['MEDV']
```

```python
X = df[['CRIM', 'ZN', 'CHAS', 'NX', 'RM', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT'
       ]]
```

```python
import statsmodels.api as sm
```

```python
x = sm.add_constant(X)
```

```python
model = sm.OLS(y,X)
print(model.fit().summary())
```
