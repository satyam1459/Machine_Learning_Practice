```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Insurance%20Premium.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
# Step3: define y and x
df.columns
```

```python
y = df['Premium']
X = df[['Age', 'Gender', 'BMI', 'Children', 'Smoker', 'Region']]
```

```python
X
```

```python
# Ordinal encoding
X.Smoker.value_counts()
```

```python
df.Smoker.replace({'yes':0,'no':1},inplace = True)
```

```python
X.Smoker.value_counts()
```

```python
X= pd.get_dummies(X)
```

```python
X
```

```python
# Split our data
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
#Select model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
#Train
model.fit(X_train,y_train)
```

```python
#predict
y_pred=model.predict(X_test)
```

```python
#evaluate
from sklearn.metrics import mean_absolute_percentage_error,r2_score
```

```python
#accuracy
mean_absolute_percentage_error(y_test,y_pred)
```

```python
#r2_score bta rha ki jitne independent var hai wo explain krrrhe premium ko
r2_score(y_test,y_pred)
```
